#Slicing (continuacion)

[inicio : stop : incremento]

Notas:
1. El stopr no es incluido
2. El incremento se puede modificar 
3. Soporta indexado alternativo (negativos)



```
frase= "Comó hacer a la gente mas positiva?, muy fácil, haz que pierdan un electrón!"
print(frase)
  #imprimir (ace) de la frase hacer
print(frase [6 : 9 : 1])


  #Valores por defecto

print(frase [ :5 :2 ])

print(len(frase))
  #incremento
print(frase [ : :2 ])
print(frase[ : : -1])

nombre= "Juan Jose"
print(nombre[1::-1])

primer=nombre[:4:]
print(primer)

print(nombre[-9::1])
```

#Métodos de la clase String (str)
En python todo es una objeto (P.O.O.), eso significa que cada tipo de dato trae consigo métodos (funciones) incluidos.

Para identificar os métodos disponibles se puede usar una de dos funciones:
1. dir ( str ) #Resumen de los métodos de esa clase
1. help ( str ) #Información completa de los métodos de esa clase 



```
nombre = "Juan José"
print(nombre)
print(dir(str))
```

Los nombres de métodos que inician con doble guión bajo, significa que son métodos privados (P.O.O), y que no tendrias por que usarlos como programador consumidor.

Los  demás métodos son públicos y de libre uso

```
print(nombre)
print( nombre.islower()) #pregunta si esta en minusculas
nombre="PEDRO FERNANDEZ"
print(nombre.capitalize()) #Primera letra con mayuscula
print(nombre.endswith("DEZ")) #como termina la frase
print(nombre.replace("E","*********"))
ind_e=nombre.find("E")#indica en donde se encuentra algo
print(ind_e)
nombre=nombre[:ind_e:]+"@"+nombre[ind_e+1::]#Remplaza
print(nombre)
print(nombre.replace("E","@",3))#Remplaza
print(nombre.lower()) #todo en minusculas

nombre = "jose"
print(nombre.upper()) #todo en mayusculas

nombre="josé josé, el principe de la cancion"
print(nombre.count("josé")) #cuenta las veces que se usa una cadena
```

```
#Método para procesar Strings

numero= " 2 , 1 , 3, 4,     5,             45,                 7  "
print(numero)

#¿Comó extraigo los numeros de la cadena?
#1.- Eliminar los espacion al principio y al final con strip()
numero=numero.strip()
print(numero)

#2.- Separo por comas
numero=numero.split(",")
print(numero)

#3.- Con un for e indices limpiar de la lista
for indice in range (len(numero)):
  numero[indice]=numero[indice].strip()
print(numero)


#4.- Convertirlos a tipo entero
for indice in range (len(numero)):
  numero[indice]= int(numero[indice])
print(numero)

#5.- usar loa numeros

print( help(str.split))
```
